```python
from datasets import load_dataset, DatasetDict
dataset  = load_dataset('HuggingFaceH4/ultrachat_200k',split=['train_sft', 'test_sft'])
dataset = DatasetDict(
    {
        "train": dataset[0],
        "test": dataset[1],
    }
)
```

```python
dataset
```

```python
from transformers import AutoTokenizer, GPT2LMHeadModel, AutoConfig, AutoModelForCausalLM
pretrained_model = "openai-community/gpt2"
model = AutoModelForCausalLM.from_pretrained(pretrained_model)
```

```python
message_contents = []
for row in dataset["train"]:
    # Extracting message contents from list of dictionaries and joining them into a single string
    message_content = " ".join([msg["content"] for msg in row["messages"]])
    message_contents.append(message_content)
```

```python
message_contents[0]
```

```python
from transformers import GPT2Tokenizer
tokenizer = AutoTokenizer.from_pretrained(pretrained_model)

max_length = 128

def tokenize_function(examples):
    return tokenizer(message_contents[:5],max_length=max_length)


# Apply the tokenization function to the entire dataset
tokenized_dataset = dataset.map(
    tokenize_function,
    batched=True,
    batch_size=10,
    remove_columns=dataset["train"].column_names
)
```

```python
tokenized_dataset
```

```python
tokenizer.pad_token = tokenizer.eos_token
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False
)
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
from transformers import Trainer, TrainingArguments

args = TrainingArguments(
    "Gpt2_Text_Generation",
   evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=5e-5,
    num_train_epochs=1,
    weight_decay=0.01,
    push_to_hub=True,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    fp16=True,
    logging_steps=50,
    # generation_max_length=128,
    # generation_num_beams=5
)

trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=args,
    data_collator=data_collator,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"]
)
```

```python
import wandb

wandb.init(project="Gpt2_Text_Generation") # Start a new wandb run

trainer.train()
```

```python
trainer.evaluate()
```
